## Task 3: Data Transformation (Any Tool)


- Rename the column `availability 365` to `days_booked`
- Convert all column names to lowercase and replace the spaces in the column names with an underscore "_".
- Remove the dollar sign and comma from the columns `price` and `service_fee`. If necessary, convert these two columns to the appropriate data type.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
# Rename the column availabilty 365 to days_booked.
df.rename(columns={'availability 365':'days_booked'}, inplace=True)
```

```python
# Convert all column names to lowercase and replace the spaces with an underscore "_"
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
# Remove the dollar sign and comma from the columns of price and service_fee. If necessary, convert these two columns to the appropriate data type.
df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)
df['service_fee'] = df['service_fee'].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

## Task 5a: Data Visualization (Any Tool)

* List the count of various room types avaliable with Airnb

* List the top 10 neighborhoods in the increasing order of their price with the help of a horizontal bar graph. Which is the cheapest neighborhood.

```python
# List the count of various room types avaliable with Airnb
sns.countplot(data=df, x='room_type')
```

```python
# Which room type adheres to more strict cancellation policy
room_policy = df[df["cancellation_policy"] == "strict"]

# Count the occurrences of each room type
room_type_counts = room_policy["room_type"]

#Visualizing using seaborn barplot
sns.barplot(x=room_type_counts.index, y=room_type_counts)
```

```python
# List the prices by neighborhood group and also mention which is the most expensive neighborhood group for rentals
neighborhood_prices = df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)

# Visualize the prices by neighborhood group using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=neighborhood_prices.index, y=neighborhood_prices)
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price')
plt.title('Average Price by Neighborhood Group')
plt.xticks(rotation=45)
plt.show()

# Mention the most expensive neighborhood group for rentals
most_expensive_neighborhood_group = neighborhood_prices.index[0]
most_expensive_neighborhood_group
```

```python
#List the top 10 neighborhoods in the increasing order of their price with the help of a horizontal bar graph. Which is the cheapest neighborhood.
# Sort the neighborhoods by price in ascending order
top_10_neighborhoods = df.groupby('neighbourhood')['price'].mean().sort_values().head(10)

# Create a horizontal bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_neighborhoods.values, y=top_10_neighborhoods.index)
plt.xlabel('Average Price')
plt.ylabel('Neighborhood')
plt.title('Top 10 Neighborhoods by Average Price')
plt.show()

# Find the cheapest neighborhood
cheapest_neighborhood = top_10_neighborhoods.index[0]
cheapest_neighborhood
```

```python
#List the neighborhoods which offer short term rentals within 10 days. Illustrate with a bar graph
short_term_rentals = df[df['days_booked'] <= 10]['neighbourhood'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=short_term_rentals.index, y=short_term_rentals.values)
plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.title('Neighborhoods with Short Term Rentals within 10 Days')
plt.xticks(rotation=45)
plt.show()
```

```python
#List the prices with respect to room type using a bar graph and also state your inferences.
# Group the data by room type and calculate the average price
room_prices = df.groupby('room_type')['price'].mean().sort_values()

# Create a bar plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=room_prices.index, y=room_prices.values)
plt.xlabel('Room Type')
plt.ylabel('Average Price')
plt.title('Average Price by Room Type')
plt.xticks(rotation=45)
plt.show()
```

```python
#Create a pie chart that shows  distribution of booked days for each neighborhood group .Which neighborhood has the highest booking percentage.
# Calculate the booking percentage for each neighborhood group
booking_percentage = df.groupby('neighbourhood_group')['days_booked'].sum() / df['days_booked'].sum() * 100

# Create a pie chart
booking_percentage.plot.pie(autopct='%1.1f%%', figsize=(8, 8))
plt.title('Booking Percentage by Neighborhood Group')
plt.ylabel('')
plt.show()

# Find the neighborhood with the highest booking percentage
highest_booking_neighborhood = booking_percentage.idxmax()
highest_booking_neighborhood
```

```python
# Using a line graph show in which year the maximum construction of rooms took place.
contruction_year_counts = df['construction_year'].value_counts().sort_values()

plt.figure(figsize=(10,6))
contruction_year_counts.plot(kind='line', marker= 'o', color='black')
plt.title('Maximum Construction of Rooms Over the Years')
plt.xlabel('Year')
plt.ylabel('No. of Rooms Constructed')
plt.grid(True)
plt.show()
```
